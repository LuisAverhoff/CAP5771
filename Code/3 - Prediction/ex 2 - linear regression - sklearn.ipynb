```python
from sklearn import datasets, linear_model
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
data = datasets.load_boston()
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df.head()
```

```python
target = pd.DataFrame(data.target, columns=["MEDV"])
target.head()
```

```python
X = df
y = target["MEDV"]
```

```python
lm = linear_model.LinearRegression()
model = lm.fit(X, y)
```

```python
predictions = lm.predict(X)
```

```python
merged = df
merged["MEDV"] = target["MEDV"]
merged.head()
```

```python
pd.plotting.scatter_matrix(merged.loc[:, ["RM", "LSTAT"]], diagonal="kde")
plt.show()
```

```python
fig, ax = plt.subplots()
scat = ax.scatter(X["RM"], X["LSTAT"], c=y, s=200, marker='o')
fig.colorbar(scat, label='MEDV: Median value of owner-occupied homes ($1000)')
plt.xlabel("RM: Average number of rooms per dwelling")
plt.ylabel("LSTAT: % lower status of the population")
plt.title("Multivariate Linear Regression")
plt.show()
```

```python
fig, ax = plt.subplots()
scat = ax.scatter(X["RM"], X["LSTAT"], c=predictions, s=200, marker='o')
fig.colorbar(scat, label='MEDV: Median value of owner-occupied homes ($1000)')
plt.xlabel("RM: Average number of rooms per dwelling")
plt.ylabel("LSTAT: % lower status of the population")
plt.title("Multivariate Linear Regression")
plt.show()
```
